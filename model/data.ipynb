```python
import pandas as pd
import numpy as np

data_df = pd.read_csv("data/hr_data.csv")

dropped_df = data_df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1)

print(f"Column Count: {len(dropped_df.columns)}")
for column in dropped_df.columns:
    print(f'Column: {column}')
    print(np.unique(dropped_df[column]))
    print("\n")
```

```python
to_categorize = ['Attrition', 'BusinessTravel', 'Department', 'Education', 'EducationField', 'EnvironmentSatisfaction', 'Gender', 'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction', 'MaritalStatus', 'OverTime', 'RelationshipSatisfaction', 'WorkLifeBalance']

value_map = []
for column in to_categorize:
    print(column)
    my_dict = {}

    unique_values = np.unique(dropped_df[column])
    sorted_values = np.sort(unique_values)
    for idx, value in enumerate(sorted_values):
        my_dict[idx] = value
        
    value_map.append({
        column: my_dict
    })

for dict_map in value_map:
    print(dict_map)
    print("\n")
```
